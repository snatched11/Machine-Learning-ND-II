```python
import math
import numpy as np
```

```python
curved = [score + 7 for score in test_scores]
print (np.mean(curved))
```

# Better Code

```python
def flat_curve (arr, n):
    return [i+n for i in arr]

def square_root (arr):
    return [math.sqrt(i) * 10 for i in arr]

test_scores = [1, 8, 2, 3]
curved = flat_curve(test_scores, 7)
curved_sqrt = square_root (test_scores)
print (np.mean(test_scores))
print (np.mean(curved))
print (np.mean(curved_sqrt))
```

```python
for score_list in test_scores, curved, curved_sqrt:
    print (np.mean(score_list))
```
